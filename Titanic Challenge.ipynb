```python
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()

train_data = pd.read_csv('../input/train.csv')

train_data['Age'].fillna(method = 'ffill',inplace = True)

test_data = pd.read_csv('../input/test.csv')
test_data['Age'].fillna(method = 'ffill',inplace = True)
```

```python
train_data.head()
```

```python
gender = {'male':1,'female':2}

train_data.Sex =[gender[item] for item in train_data.Sex]

train_data.head()

```

```python
train_labels = train_data['Survived']

train_features = train_data.drop(['Survived'],axis = 1)
train_features = train_features.drop(['Name'],axis = 1)
train_features = train_features.drop(['Ticket'],axis = 1)
train_features = train_features.drop(['Cabin'],axis = 1)
```

```python
test_features = test_data.drop(['Name'],axis = 1)
test_features = test_features.drop(['Ticket'],axis = 1)
test_features = test_features.drop(['Cabin'],axis = 1)
test_features['Embarked'] = test_features['Embarked'].fillna('S')
test_features.Embarked =[embarked[item]for item in test_features.Embarked]
test_features.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(train_features,train_labels)

results = knn.predict(test_features)
```
